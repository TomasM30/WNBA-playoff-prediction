## Coaches

We will take a look at the coaches data first.


```python
coaches.head()
```

The `coaches.csv` dataset contains the following columns:

- `coachID`: unique identifier for each coach. It is also present in the `awards_players.csv`dataset. 
- `year`: the year the coach was active.
- `tmID`: the team the coach was coaching.
- `lgID`: the league the team was in.
- `stint`: the stint of the coach in the team.
- `won`: the number of games won by the team.
- `lost`: the number of games lost by the team.
- `post_wins`: the number of post-season wins by the team.
- `post_losses`: the number of post-season losses by the team.


This dataset contains information about the performance of the coaches in the NBA, including the number of games won and lost by the team, as well as the number of post-season wins and losses. It also says if the coach coached more than one team in the same year.
